```python
import pandas as pd
```

```python
excel_path = "data/april_generator2026.xlsx"
df = pd.read_excel(io = excel_path, sheet_name = "Operating",header =2)
```

```python
df.shape
```

```python
df.columns
```

```python
cols = ['Plant ID','Plant Name','Plant State','County','Balancing Authority Code','Generator ID','Nameplate Capacity (MW)','Energy Source Code','Operating Month','Operating Year','Latitude',
       'Longitude']
df_filtered = df[cols]
df_filtered.shape
```

```python
# filter the balancing authority in {'MISO', 'SWPP', 'SPA'} and energy source code in {'WND','SUN}
df_new = df_filtered[
    (df_filtered['Balancing Authority Code'].isin(['MISO', 'SWPP', 'SPA'])) & 
    (df_filtered['Energy Source Code'].isin(['WND','SUN']))
    ]
print(df_new.shape)
df_new['Energy Source Code'].value_counts()
```

```python
# change the datatype into not nullable integer type
df_new = df_new.astype({'Plant ID': 'Int64','Operating Month': 'Int64','Operating Year': 'Int64'})

df_new[['Plant ID', 'Operating Month', 'Operating Year']].dtypes
```

```python
df_new[['Plant ID', 'Operating Month', 'Operating Year']].isna().sum()
```

```python
# change dataframe into JSON 
json_path = "data/nodes.json"
df_new.to_json(json_path, orient='records',indent =2)
```
